```python
import math
import numpy as np
import matplotlib.pyplot as plt

x = np.array(range(100))
y = np.sqrt(x)
z=[]

for i in range(0,len(x),1):
  z.append(math.pow(x[i],1.0/3.0))

plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
